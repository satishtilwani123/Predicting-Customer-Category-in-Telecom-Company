```python
#Lets Select Features Based on Mutual Information.
MI = []
for i in df.columns[:-1]:
    mi = MIC(df[[i]], df['custcat'])
    MI.append({'Feature':i, 'MI-Score':mi[0]})
```

### Start Building Model

```python
def TrainingKNN(X, Y, k):
    
    ss = StandardScaler()
    knn = KNeighborsClassifier(k)
    
    #Normalization of Data
    X_Normalized = ss.fit_transform(X)
    #Data Splitting
    train_x, test_x, train_y, test_y = train_test_split(X_Normalized, Y, test_size=0.2, random_state=0)
    
    #Train Model
    knn.fit(train_x, train_y)
    
    #Predictions
    train_predict = knn.predict(train_x) 
    test_predict = knn.predict(test_x)
    
    #Measuring Accuracy of Model
    train_acc = accuracy_score(train_y, train_predict)
    test_acc = accuracy_score(test_y, test_predict)
    
    return train_acc, test_acc
```

```python
X = df[['tenure', 'age', 'income', 'ed', 'employ', 'gender']]
Y = df['custcat']
```

```python
k = 3
```

```python
train_acc, test_acc = TrainingKNN(X, Y, k)
```

```python
KNN_Results = []
for i in range(1, 10):
    train_acc, test_acc = TrainingKNN(X, Y, i)
    KNN_Results.append({'k':i, 'Train-accracy':train_acc, 'Test-accuracy':test_acc})
```

```python
Results = pd.DataFrame.from_records(KNN_Results)
```

```python
#The best accuracy we got so far on testing data is 0.365 with k = 9.
Results.sort_values(by='Test-accuracy')
```

```python
Results.set_index('k', inplace=True)
```

```python
Results.plot(kind='line', figsize=(10,8))

plt.title('Plotting Train & Test Accuracy Based on K values')

plt.xlabel('K')
plt.ylabel('Accuracy')
plt.show()
plt.close()
```
